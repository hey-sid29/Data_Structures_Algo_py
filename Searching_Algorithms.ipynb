#### Creating a  Linear Search Algorithm

```python
#Going back to the function we created: locate_card(cards, query_card)


def locate_card(cards, query_card):


    print("List of cards: ", cards)
    print("Queried card: ", query_card)


    #Creating a variable which controls the position = 0 
    pos = 0

    #Creating a loop for repetition
    while True:
        
        print("position: ", pos)
        
        #Checking for the condition, if true return the position where the card has been found
        if (cards[pos] == query_card):
            return pos


        #if the the value wasn't found increasing the value of pos by 1
        pos+=1

        #Check if the list has ended and position points to the length of the list
        if (pos == len(cards)):
            return -1


#This is an  implementation of Linear Search Algorithm


```

### LinearSearch solution in plain language: 
<br>
1) Start from ""position = 0"" <br>
2) Check if ""card[position] == query"" <br>
3) If condition => True, return the value of ""position"" <br>
4) If condition => False, increment value of ""position by 1"" and repeat above steps. <br>
5) If query was not found ""return -1""

### BinarySearch Algorithm:

```python
# 1) Find the middle element of the list.
# 2) If it matches with the queried number, return the middle index.
# 3) If it is less than queried number, then search the first half of the list. 
# 4) If it is greater than the queried number, then search the second half of the list. 
# 5) If none element match, return -1



def bin_locate_card(cards, query_card):
    lo = 0; hi = len(cards) - 1

    while lo<=hi:
        mid = (hi+lo)//2
        mid_num = cards[mid]

        print("low: ", lo, "\n", "high: ",hi, "\n", "mid index: ", mid, "\n", "mid number: ", mid_num )

        if mid_num == query_card:
            return mid

        elif mid_num < query_card:
            hi = mid - 1

        elif mid_num > query_card:
            lo = mid + 1

    return -1
```

```python
#2 does occur on pos = 8, before the actual output = 9, our function gets confused and returns any value that matches to query
# to fix it: we can check if the number is the first occurence in the list or not

def test_location(cards, query_card, mid):
    mid_num = cards[mid]
    print("Mid: ", mid, ", ", "mid number: ", mid_num)
    if mid_num == query_card:
        if mid-1>=0 and cards[mid-1] == query_card:
            return 'left'
        else: 
            return 'found'
    elif mid_num<query_card:
        return 'left'
    else: 
        return 'right' 


def bin_locate_card(cards, query_card):
    lo, hi = 0, len(cards)-1
    while lo <= hi:
        print('low: ', lo, ', ', 'high: ', hi)
        mid = (lo+hi)//2
        
        result = test_location(cards, query_card, mid)

        if result == 'found':
            return mid

        elif result == 'left':
            hi = mid - 1
        
        elif result == 'right':
            lo = mid + 1

            
    return -1
```

### Binary_Search Vs. Linear_Search: which is effecient:

```python
large_test_case = {
    'input' : {
        'cards' : list(range(10000000, 0, -1)),
        'query_card' : 2
    },
    'output' : 9999998
    
}
```

```python
#Linear Search Result on a large array
result, passed, runtime = evaluate_test_case(locate_card, large_test_case, display=False)
print("Result: {} \nPassed: {} \nExecution_Time: {}ms".format(result, passed, runtime))
```

```python
#Binary Search Result on a large array
result, passed, runtime = evaluate_test_case(bin_locate_card, large_test_case, display=False)
print("Result: {} \nPassed: {} \nExecution_Time: {}ms".format(result, passed, runtime))
```

Linear_Search_Time: 1023.65ms <br>
Binary_Search_time: 0.647ms <br>
So we see binary search is far more time efficient than Linear Search even in such a corner case
